```python
# Cell 1 — Multi-Agent Conversation: Father & Son
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv(r'C:\Users\Gurveer\ds-portfolio\.env')
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

# --- Role instructions ---
father_role = (
    "You are a frustrated father. Your son keeps parking his car "
    "without regard to other people in the house, and they have a hard "
    "time parking their cars. Be firm but respectful."
)
son_role = (
    "You are a polite and empathetic son. Acknowledge the dad's "
    "frustration, ask clarifying questions, and offer solutions. "
    "Keep your tone constructive."
)

# --- Shared conversation history ---
conversation = [
    {"role": "system", 
     "content": "This is a dialogue simulation between a father and a son."},
]

# --- Helper: one agent speaks with its own system instruction ---
def get_reply(role_instruction, conversation, model="gpt-4o-mini"):
    resp = client.chat.completions.create(
        model=model,
        messages=[{"role": "system", 
                   "content": role_instruction}] + conversation,
    )
    return resp.choices[0].message.content

# --- Start the conversation with the father ---
father_msg = "Hello, the way you park your car, I cannot fit my car in the driveway."
conversation.append({"role": "user", "content": father_msg})
print("Dad:", father_msg)

# --- Run turns back and forth ---
for _ in range(4):
    son_msg = get_reply(son_role, conversation)
    print("\nSon:", son_msg)
    conversation.append({"role": "assistant", "content": son_msg})

    father_msg = get_reply(father_role, conversation)
    print("\nDad:", father_msg)
    conversation.append({"role": "user", "content": father_msg})
```
